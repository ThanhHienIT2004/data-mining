```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

```

```python
path = "C:/DoAn/merged_movies.csv"
df = pd.read_csv(path, encoding="latin1")

df.head()
```

```python
df.info()
```

```python
df_cluster = df[['Plot']]
df_cluster.head()
```

```python
df = df.dropna()
df.shape
```

```python
tfidf = TfidfVectorizer(
    stop_words="english",
    max_features=2000
)

X = tfidf.fit_transform(df['Plot'])

X.shape
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X)

df['Cluster'] = clusters
df.head()
```

```python
for i in range(5):
    print(f"\n===== CLUSTER {i} =====")
    print(df[df['Cluster'] == i]['Plot'].head(3).values)
```

```python
df['Cluster'].value_counts()
```

```python
output_path = "C:/DoAn/movies_clustered.csv"
df.to_csv(output_path, index=False, encoding="utf-8-sig")

print("Đã xuất file:", output_path)
```
