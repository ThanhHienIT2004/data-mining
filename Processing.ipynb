# Feature Engineering for Movie Recommendation System

Notebook này thực hiện:
- Xử lý đặc trưng cho hệ gợi ý phim
- Kết hợp nội dung phim (Plot) và thể loại (Genre)
- Chuẩn bị dữ liệu cho bước vector hóa


```python
import pandas as pd
```

```python
model_df = pd.read_csv("movies_model_Preprocessing.csv")

# Genre_list đang ở dạng string → chuyển lại thành list
import ast
model_df["Genre_list"] = model_df["Genre_list"].apply(ast.literal_eval)

model_df.head()
```

```python
model_df.info()
```

```python
# Chuyển list genre thành chuỗi
model_df["Genre_text"] = model_df["Genre_list"].apply(
    lambda genres: " ".join(genres) if isinstance(genres, list) else ""
)

model_df[["Genre_list", "Genre_text"]].head()
```

```python
GENRE_WEIGHT = 2  # tăng ảnh hưởng thể loại

model_df["combined_text"] = model_df.apply(
    lambda row: row["Plot_clean"] + " " + (" " + row["Genre_text"]) * GENRE_WEIGHT,
    axis=1
)

model_df[["Title", "combined_text"]].head()
```

```python
model_df["combined_length"] = model_df["combined_text"].apply(
    lambda x: len(x.split())
)

model_df["combined_length"].describe()
```

```python
model_df.to_csv("movies_processing.csv", index=False)

print("✅ Feature engineering hoàn tất & đã lưu file")
```
